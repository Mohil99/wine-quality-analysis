# Task 5: Visualizing Distribution of Key Variables with Boxplots and Histograms
## Goal: Explore relationships between wine quality and several key chemical features.
# Task:
## Visualize the distribution of key variables (alcohol, sulphates, citric acid, and fixed_acidity) with boxplots and histograms.
## What do you observe about how wine quality is related to these features? Are there any patterns that stand out? note: Create a subplot showing boxplots and histograms for each of the selected features (alcohol, sulphates, citric_acid, and fixed_acidity) with a title describing the feature being visualized.

```python
df.describe().T
```

```python
chemical_features = ["alcohol",
                     "sulphates",
                     "citric acid",
                     "fixed acidity"
                    ]
```

```python
plt.figure(figsize=(15,12))

for i, feature in enumerate(chemical_features, 1):
    
    
#for boxplot
    plt.subplot(2,4,i)
    sns.boxplot(x=df[feature])
    plt.title(f"Boxplot of {feature}")

#for histogram
    plt.subplot(2,4, i+4)
    sns.histplot(df[feature], kde=True, bins = 15 , color ="green")
    plt.title(f"Histogram of {feature}")

plt.show()
```

# Task 6: Scatter Plot for Fixed Acidity vs Alcohol Content
## Goal: Investigate the relationship between acidity and alcohol content.
# Task:
## Create scatter plots to analyze the relationship between fixed_acidity and alcohol , as well as volatile_acidity and alcohol .
## Do you observe any trends or patterns? How do these variables appear to correlate with each other?

```python
facidity = df["fixed acidity"]
vacidity = df["volatile acidity"]
alcohol_content = df["alcohol"]
```

```python
plt.scatter(facidity,alcohol_content)
plt.scatter(vacidity,alcohol_content)

plt.title("alcohol vs fixed acidity and volatile acidity")
plt.xlabel("alcohol content")
plt.legend(["Fixed Aciity","Volatile Acidity"])
plt.show()
```

# Do you observe any trends or patterns? How do these variables appear to correlate with each other?

*Since whenever alcohol content increases not necessarily fixed acidity and volatile acidity also increases*
